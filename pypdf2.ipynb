```python
#parsing the pdf file
import spacy
from PyPDF2 import PdfReader
nlp = spacy.load("en_core_web_md")
file_path = r"C:\Users\VISHWA 'S DELL\OneDrive\Desktop\NLP1\NIST-Cybersecurity-Framework-Policy-Template-Guide-v2111Online.pdf"
reader= PdfReader(file_path)
#extracting text from all pages
text=""
for page in reader.pages:
    text +=page.extract_text()
#processing the text
doc=nlp(text)

for ent in doc.ents:
    print(ent.text,ent.label_)
for token in doc:
    print(token.text,token.pos_,token.dep_,token.head.text)
```

```python
#export to json
import json
output_path= r"C:\Users\VISHWA 'S DELL\OneDrive\Desktop\NLP1\output.json"
with open(output_path, 'w') as f:
    json.dump([{"text": ent.text, "label": ent.label_} for ent in doc.ents], f, indent=4)
    print("Entities exported to JSON file successfully.")
    
```

```python
#check if the file exists in json format
if os.path.exists(output_path):
    print(f" File exported: {output_path}")
    
    # Read and display a sample
    with open(output_path, "r", encoding="utf-8") as f:
     data = json.load(f)
     print(" Sample Named Entity:", data[0])
     print(" Sample POS Tag:", data[0])
else:
    print("Export failed.")
```
